```python
class NB_true:
    """
    NAIVE BAYES CLASS AND FUNCTIONS
    
    FAIR USE: credit to Jonathan Hendriks 
    """
    def fit(self, data, cls):
        self.cls = cls
        self.k = np.unique(cls).size
        self.d,self.N = data.shape
        self.pi = np.zeros((self.k,))

        self.u = np.zeros((self.d,self.k))
        self.S = np.zeros((self.d,self.k))
       
        for j in range(0, self.k):
            self.pi[j] = np.sum(cls==j)/N
            for dim in range(0, self.d):
                obs = data[dim, cls == j]
                self.u[dim,j] = np.mean(obs)
                self.S[dim, j] = np.var(obs)     
                        
    def pxc(self, x, mean, var1):
        """
        1/sqrt(2*pi*Eps)*exp{ -1/2*(x-u).T*(Eps^-1)*(x-u) }
        """
        dif = (x-mean) 
        e = np.exp((-1/2)*(dif.T)*(dif/var1))
        
        return (1/(math.sqrt(2*math.pi*var1))*e)

    def predict(self, data):
        post = np.ones((self.N, self.k))
        y = np.zeros((self.N,))
        
        for n in range(self.N):
            for j in range(self.k):
                for dim in range(self.d):
                    post[n, j] *= self.pxc(data[dim, n], self.u[dim, j], self.S[dim, j])
                post[n, j] *= self.pi[j]                                       
            y[n] = np.argmax(post[n,:])  

        return y
```

# Implement a Logistic Regression Classifier

Implement your own Logistic Regression classifier (from first principles). It should be generic and be able to work on $d$ dimensional data (and two classes). Please make your implementation as modular as possible, having a seperate function for the Hessian and the gradient vector (write it in a seperate python file which you can import into the notebook). A regularization term based on a Gaussian prior (with zero mean and covariance matrix $\lambda\mathbf{I}$) must be included.  Optionally, a bias term also needs to be incorporated into your classifier - this can be implemented by augmenting the training dataset with an additional all-one feature.

```python
class LR:
    """
    LGISTIC REGRESSION CLASS AND FUNCTIONS
    """
    def fiterate(self, data, classes):
        """
        w_k+1 = w_k - inv(hessian(w_k))*grad_l(w_k)
        until abs((w_k+1 - w_k)/w_k) < 1e-5
        """
        self.lam = 2
        
        self.y = classes
        
        self.N = data.shape[1]
        ones = np.ones((1, self.N))
        self.X = np.vstack((ones, data))
        self.dim = (self.X).shape[0]
        
        self.w_k = np.zeros(shape = (self.dim,))
        w_kn = np.ones(shape = (self.dim,))
        check = np.zeros(shape = (self.dim,))
        check = 1
        while check >= 1e-5:
            print("ITERATION")
            hess = self.hessian(self.w_k)
            inv_hess = np.linalg.inv(hess)
            l = self.grad_l()
            ihl = np.reshape(inv_hess.dot(l), newshape = (3,))
            check = abs(np.sum(ihl))
            print("Inv_HESS")
            print(inv_hess)
            print("GRAD L")
            print(l)
            print("HESS.dot(GRAD L)")
            print(ihl)
            print("")
            w_kn = self.w_k - ihl
            self.w_k = w_kn
                

    def sig(self, a):
        return (1/(1+math.exp(-a)))            
        
        
    def hessian(self, w):
        """
        H(w) = sum_N{ sig(w.T*x_n)(1-sig(w.T*x_n))(x_n*x_n.T) + (1/lambda)*I }
        """
        h = np.zeros(shape = (self.dim, self.dim))
        
        for n in range(0, self.N):
            a = np.dot(w, self.X[:,n])
            b = self.sig(a)*(1-self.sig(a))
            c = np.outer(self.X[:,n],self.X[:,n])
            h = h+b*c
        h += np.eye(self.dim, self.dim)/self.lam
        return h
        
        
    def grad_l(self):
        """
        grad_l(w_j) = sum_N{ (sigm(w.T*x) - y[n])*x[n]
        """
        gl = np.zeros(shape = (self.dim,))
        c = np.zeros(shape = (self.N,))
        
        for n in np.array(range(0, self.N)):
            a = (self.w_k.T).dot(self.X[:,n])
            
            if self.y[n] == 0:
                c[n] = 1
            else:
                c[n] = 0
                
            b = self.X[:, n].dot(self.sig(a)-c[n])
            gl += b
            
        gl += (1/self.lam)*self.w_k
        
        return gl[:,np.newaxis]

    
    def predict(self, data):
        """
        p(c1|x,w) = sigm(w.T*x)
        """
        y = np.ones(shape = (self.N,))
        
        for n in range(0,self.N):
            a = (self.w_k.T).dot(self.X[:,n])
            pc1 = self.sig(a)
            if pc1 >= 0.5:
                y[n] = 0
        return y
            
```

```python
def sig(a):
    return (1/(1+math.exp(-a)))  

def grad_l(X,w_k,y):
    """
    grad_l(w_j) = sum_N{ (sigm(w.T*x) - y[n])*x[n]
    """
    dim,N = X.shape
    lam = 1
    gl = np.zeros(shape = (dim,))
    c = np.zeros(shape = (N,))

    for n in np.array(range(0, N)):
        a = (w_k.T).dot(X[:,n])

        if y[n] == 0:
            c[n] = 1
        else:
            c[n] = 0

        b = X[:, n].dot(sig(a)-c[n])
        gl += b

    gl += (1/lam)*w_k

    return gl[:,np.newaxis]

def hessian(X,w):
        """
        H(w) = sum_N{ sig(w.T*x_n)(1-sig(w.T*x_n))(x_n*x_n.T) + (1/lambda)*I }
        """
        dim,N = X.shape
        lam = 1
        h = np.zeros(shape = (dim, dim))
        
        for n in range(0, N):
            a = np.dot(w, X[:,n])
            b = sig(a)*(1-sig(a))
            c = np.outer(X[:,n],X[:,n])
            h = h+b*c
        h += np.eye(dim, dim)/lam
        return h
```
